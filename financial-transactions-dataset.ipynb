## About the Dataset

### Overview

This dataset is a comprehensive financial dataset provided by a banking institution, combining **transaction records, customer demographics, card information, and fraud labels** across the **2010s decade**.

It is designed to support a wide range of analytical and machine learning use cases, including **fraud detection**, **customer behavior analysis**, and **financial forecasting**. The dataset structure closely mirrors real-world banking data, making it ideal for building and testing production-grade data pipelines and analytical models.

---

## Dataset Components

### 1. Transaction Data (`transactions_data.csv`)
- Detailed transaction-level records including:
  - Transaction amounts
  - Timestamps
  - Merchant details and transaction types
- Covers transactions throughout the 2010s
- Suitable for spending pattern analysis, time-series analysis, and fraud detection model development

---

### 2. Card Information (`cards_data.csv`)
- Credit and debit card metadata including:
  - Card types
  - Spending limits
  - Activation dates
- Linked to customer accounts via `card_id`
- Enables deeper insight into customer financial behavior and risk exposure

---

### 3. Merchant Category Codes (`mcc_codes.json`)
- Standardized Merchant Category Codes (MCC)
- Provides industry-level classification for merchants
- Enables transaction categorization and merchant-level spending analysis

---

### 4. Fraud Labels (`train_fraud_labels.json`)
- Binary classification labels for transactions
- Indicates whether a transaction is fraudulent or legitimate
- Designed for supervised learning and fraud detection model training

---

### 5. User Data (`users_data.csv`)
- Customer demographic and account-related information
- Enables customer segmentation, behavioral analysis, and personalized analytics

---
